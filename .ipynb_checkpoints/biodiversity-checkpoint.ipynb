```python
# Looks like we have the data of the species where we are given their scientific name,
# common name and their endangered status, the last one is mostly NaN, 
# possibly it is because these species aren`t endangered

duplicatesSpecies = dfSpecies[dfSpecies['scientific_name'].duplicated(keep=False)]
```

```python
duplicatesSpecies = duplicatesSpecies.sort_values(by='scientific_name', ascending=True)
duplicatesSpecies
```

```python
# Common names are repeating on multiple species, different plants that are similiar to each other would be called
# same common name, similarily some plants have more than one common name, that created duplicates on scientific 
# names column

# I see that dropping the duplicates on columns scientific_name and common_name is not going to change what we want
# to describe here, actually it will make the data more acurate as the duplicates would make the result not true
# with the species that are not endagered being repeated more often.

dfSpecies_copy = dfSpecies.copy()
dfSpecies_copy.drop_duplicates(['scientific_name'], inplace=True, keep='first')
dfSpecies_copy.drop_duplicates(['common_names'], inplace=True, keep='first')
```

```python
dfSpecies_copy.conservation_status.fillna('Not Endangered', inplace=True)
```

```python
dfSpecies_copy['category'].value_counts().plot(kind='bar', figsize=(10, 6))
plt.xlabel('Category')
plt.ylabel('Count')
plt.title('Distribution of Categories')
plt.show()
```

```python
dfHeatMap = dfSpecies_copy[['category', 'conservation_status']]

# create a frequency table
freq_table = pd.crosstab(index=dfSpecies_copy['category'], columns=dfSpecies_copy['conservation_status'])

single_val_perc = 100 / 5229

#change the values in to percentages to avoid very large numbers
def perc(val):
    return val * single_val_perc


freq_table_perc = freq_table.apply(perc)

plt.figure(figsize=(10,7))

# create a heatmap
sns.heatmap(freq_table_perc, annot=True, cmap="YlGnBu")

plt.title('Heatmap of Category vs Conservation Status')
plt.xlabel('Conservation Status')
plt.ylabel('Category')
plt.show()
```

```python
# This graphic does not show the true data, there is much more plants in DF, will be better to change 
# each of the categories in to a equal value, can do it by creating the inference table
```

```python
# create an inference table by converting frequencies to proportions within each category
inference_table = freq_table.div(freq_table.sum(axis=1), axis=0)

plt.figure(figsize=(10,7))

# create a heatmap
sns.heatmap(inference_table, annot=True, cmap="YlGnBu", fmt=".2f")

plt.title('Heatmap of Category vs Conservation Status (Proportions)')
plt.xlabel('Conservation Status')
plt.ylabel('Category')
plt.show()
```
